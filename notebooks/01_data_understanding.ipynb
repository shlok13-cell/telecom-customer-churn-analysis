```python
missing = train.isnull().sum()

missing_percent = (missing / len(train))*100

pd.DataFrame({
    'Missing_Count':missing,
    'Missing_Percentage':missing_percent
}).sort_values(
    by='Missing_Percentage',
    ascending=False
)
```

```python
train['NetworkScore'] = train['NetworkScore'].fillna(
    train['NetworkScore'].median()
)

train['Age'] = train['Age'].fillna(
    train['Age'].median()
)

train['IsActiveMember'] = train['IsActiveMember'].fillna(
    train['IsActiveMember'].mode()[0]
)

train['EstimatedMonthlyUsage'] = \
train['EstimatedMonthlyUsage'].fillna(
train['EstimatedMonthlyUsage'].median()
)
```
